Суть этой функции в том, что запускается 3 цикла одновременно. Первый цикл запускает опреации пользователя по порядку. Второй цикл запускает опреации пользователя по порядку начиная с первого элемента. Третий цикл показывает номер операции второго цикла. Далее переменная k - это время работы операции o. Далее идёт отбор при котором проверяется в какой списк нужно записать переменною k.Также из расчета выкидываются такие k которые равны 1200 секунд (20 минут).
